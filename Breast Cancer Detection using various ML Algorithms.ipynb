# BreastCancerDetection using various ML Algorithms

### *Importing Libraries*

```python
import pandas as pd #useful for loading the dataset
import numpy as np #to perform array
from matplotlib import pyplot
```

### *Load Dataset*

```python
dataset = pd.read_csv('data.csv') #Laoding the dataset
```

### *Summarize Dataset*

```python
print(dataset.shape)
print(dataset.head(5))
print(dataset.describe())
```

### *Mapping Class String Values to Numbers*

```python
#Mapping B-0 and M-1

dataset["diagnosis"] = dataset["diagnosis"].map({"B": 0, "M": 1}).astype(int)
print(dataset.head)
```

### Segregate Dataset into X & Y

```python
#Independent Variable

X = dataset.iloc[:, 2:32].values # Column 2 to column 32 as X
X
```

```python
# Dependent Variable
Y = dataset.iloc[:,1].values
Y
```

### *Splitting Dataset into Train & Test*

```python
#Test and Train
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
```

### *Feature Scaling*

```python
# we scale our data to make all the features contribute equally to the result

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train) #fit method is calculating the mean and variance of each of the features present in our data
X_test = sc.transform(X_test) #Transform method is transforming all the features using the respective mean and variance

#We want our test data to be a completely new and a surprise set for our model
```

# Model Score 
### *Validating some ML algorithms by its accuracy 

```python
# We are importing 6 ML Algorithms

# Linear Discriminant Analysis
# Logistic Regression
# DecisionTree Classifier
# K Nearest Neighbors (KNN)
# Gaussian NB
# Support Vector Machine

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold
```

```python
#Appending all the models

models = []
models.append(("LR", LogisticRegression(solver="liblinear", multi_class="ovr")))
models.append(("LDA", LinearDiscriminantAnalysis()))
models.append(("KNN", KNeighborsClassifier()))
models.append(("DTC", DecisionTreeClassifier()))
models.append(("GNB", GaussianNB()))
models.append(("SVM", SVC(gamma="auto")))
```

```python
results = []
names = []
res = []
for name, model in models:
    kfold = StratifiedKFold(n_splits=10, random_state=None)
    cv_results = cross_val_score(model, X_train, y_train, cv=kfold, scoring="accuracy")
    results.append(cv_results)
    names.append(name)
    res.append(cv_results.mean())
    print('%s: %f' % (name, cv_results.mean()))

pyplot.ylim(.900, 1.000)
pyplot.bar(names, res, color ="blue", width = 0.7)

pyplot.title("Algorithm Comparison")
pyplot.show()
```
